```python
defi_cat = {
    'Applied Learning' : {'full' : 0.06, 'exp': 0.08},
    'Literacy & Language' : {'full' : 0.4, 'exp': 0.4 },
    'Math & Science' : {'full' : 0.27, 'exp': 0.27 },
    'Music & The Arts' : {'full' : 0.09, 'exp': 0.09 },
    'Health & Sports' : {'full' : 0.08, 'exp': 0.07 },
    'Special Needs' : {'full' : 0.05, 'exp': 0.06 },
    'History & Civics' : {'full' : 0.03, 'exp': 0.03 },
    'Warmth Care & Hunger' : {'full' : 0.02, 'exp': 0 },
}

counter = -1
def replace_cat(row):
    global counter
    counter +=1
    if ',' in str(row):
        value = str(row).split(',')
        value[0] = str(value[0]).strip()
        value[1] = str(value[1]).strip()
        value_of_prob_1 = 0
        value_of_prob_2 = 0
        if 'Fully' in str(projects_cleaned.iloc[counter]['Project Current Status']) :
            value_of_prob_1 = defi_cat[str(value[0])]['full']
            value_of_prob_2 = defi_cat[str(value[1])]['full']
            if value_of_prob_1 > value_of_prob_2 :
                return value[0]
            else :
                return value[1]

        else :
            value_of_prob_1 = defi_cat[str(value[0])]['exp']
            value_of_prob_2 = defi_cat[str(value[1])]['exp']
            if value_of_prob_1 > value_of_prob_2 :
                return value[0]
            else :
                return value[1]

    else :
        return str(row)
```

```python
## Chi2 Test
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2_contingency

class ChiSquare:
    def __init__(self, dataframe):
        self.df = dataframe
        self.p = None #P-Value
        self.chi2 = None #Chi Test Statistic
        self.dof = None
        
        self.dfObserved = None
        self.dfExpected = None
        
    def _print_chisquare_result(self, colX, alpha):
        result = ""
        if self.p<alpha:
            result="{0} is IMPORTANT for Prediction".format(colX)
        else:
            result="{0} is NOT an important predictor. (Discard {0} from model)".format(colX)

        print(result)
        
    def TestIndependence(self,colX,colY, alpha=0.01):
        X = self.df[colX].astype(str)
        Y = self.df[colY].astype(str)
        
        self.dfObserved = pd.crosstab(Y,X) 
        chi2, p, dof, expected = stats.chi2_contingency(self.dfObserved.values)
        self.p = p
        self.chi2 = chi2
        self.dof = dof 
        
        self.dfExpected = pd.DataFrame(expected, columns=self.dfObserved.columns, index = self.dfObserved.index)
        
        self._print_chisquare_result(colX,alpha)

df = projects_cleaned.copy()
df['dummyCat'] = np.random.choice([0, 1], size=(len(df),), p=[0.25, 0.75])

#Initialize ChiSquare Class
cT = ChiSquare(df)

#Feature Selection
testColumns = ['Project Type','Project Subject Category Tree','Project Grade Level Category','Project Resource Category','Project Posted Month','dummyCat']
for var in testColumns:
    cT.TestIndependence(colX=var,colY="Project Current Status" ) 
```

```python
## Chi2 Test for all features
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2_contingency

class ChiSquare:
    def __init__(self, dataframe):
        self.df = dataframe
        self.p = None #P-Value
        self.chi2 = None #Chi Test Statistic
        self.dof = None
        
        self.dfObserved = None
        self.dfExpected = None
        
    def _print_chisquare_result(self, colX, alpha):
        result = ""
        if self.p<alpha:
            result="{0} is IMPORTANT for Prediction".format(colX)
        else:
            result="{0} is NOT an important predictor. (Discard {0} from model)".format(colX)

        print(result)
        
    def TestIndependence(self,colX,colY, alpha=0.01):
        X = self.df[colX].astype(str)
        Y = self.df[colY].astype(str)
        
        self.dfObserved = pd.crosstab(Y,X) 
        chi2, p, dof, expected = stats.chi2_contingency(self.dfObserved.values)
        self.p = p
        self.chi2 = chi2
        self.dof = dof 
        
        self.dfExpected = pd.DataFrame(expected, columns=self.dfObserved.columns, index = self.dfObserved.index)
        
        self._print_chisquare_result(colX,alpha)

df1 = df.copy()
df1['dummyCat'] = np.random.choice([0, 1], size=(len(df1),), p=[0.25, 0.75])

#Initialize ChiSquare Class
cT = ChiSquare(df1)

#Feature Selection
testColumns = list([a for a in df1.columns if a != 'Project Current Status']) 
#['Project Type','Project Subject Category Tree','Project Grade Level Category','Project Resource Category','Project Posted Month','dummyCat']
for var in testColumns:
    cT.TestIndependence(colX=var,colY="Project Current Status" ) 
```

```python
## modeling
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from imblearn.pipeline import make_pipeline as make_pipeline_imb
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import NearMiss
from imblearn.metrics import classification_report_imbalanced
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
def print_results(headline, true_value, pred):
    print(headline)
    print("accuracy: {}".format(accuracy_score(true_value, pred)))
    print("precision: {}".format(precision_score(true_value, pred)))
    print("recall: {}".format(recall_score(true_value, pred)))
    print("f1: {}".format(f1_score(true_value, pred)))
```

```python
# our classifier to use
#classifier = LogisticRegression
classifier= RandomForestClassifier
```

```python
data = df.loc[:, df.columns != 'Project Current Status']
target = df['Project Current Status']
```

```python
# splitting data into training and test set
X_train, X_test, y_train, y_test = train_test_split(data, target,test_size=0.2, random_state=42)
```

```python
# build normal model
pipeline = make_pipeline(classifier(random_state=42))
model = pipeline.fit(X_train, y_train)
prediction = model.predict(X_test)
```

```python
# build model with SMOTE imblearn
smote_pipeline = make_pipeline_imb(SMOTE(random_state=4), classifier(random_state=42))
smote_model = smote_pipeline.fit(X_train, y_train)
smote_prediction = smote_model.predict(X_test)
```

```python
# build model with undersampling
nearmiss_pipeline = make_pipeline_imb(NearMiss(random_state=42), classifier(random_state=42))
nearmiss_model = nearmiss_pipeline.fit(X_train, y_train)
nearmiss_prediction = nearmiss_model.predict(X_test)
```

```python
# print information about both models
print()
print("normal data distribution: {}".format(Counter(target)))
X_smote, y_smote = SMOTE().fit_sample(data, target)
print("SMOTE data distribution: {}".format(Counter(y_smote)))
X_nearmiss, y_nearmiss = NearMiss().fit_sample(data, target)
print("NearMiss data distribution: {}".format(Counter(y_nearmiss)))
```

```python
# classification report
print(classification_report(y_test, prediction))
print(classification_report_imbalanced(y_test, smote_prediction))
print(classification_report_imbalanced(y_test,nearmiss_prediction ))
```

```python
print()
print('normal Pipeline Score {}'.format(pipeline.score(X_test, y_test)))
print('SMOTE Pipeline Score {}'.format(smote_pipeline.score(X_test, y_test)))
print('NearMiss Pipeline Score {}'.format(nearmiss_pipeline.score(X_test, y_test)))
```

```python
print()
print_results("normal classification", y_test, prediction)
print()
print_results("SMOTE classification", y_test, smote_prediction)
print()
print_results("NearMiss classification", y_test, nearmiss_prediction)
```
